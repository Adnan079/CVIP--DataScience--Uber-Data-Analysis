# Data Visualization

```python
uber["CATEGORY"].value_counts().plot(kind = "bar", color = "brown", figsize=(6,4))
mp.title("Distribution of trips according to category")
mp.xlabel("CATEGORY")
mp.xticks(fontsize = 12)
mp.ylabel("NUMBER OF TRIPS")
mp.show()
```

```python
uber["PURPOSE"].value_counts().plot(kind = "bar", color = "GREY", figsize=(12,5))
mp.title("Distribution of trips according to purpose")
mp.xlabel("PURPOSE")
mp.xticks(fontsize = 12)
mp.ylabel("NUMBER OF TRIPS")
mp.show()
```

```python
uber['MONTH*'] = pd.DatetimeIndex(uber['START_DATE']).month
```

```python
dic = {1:'Jan', 2: 'Feb', 3: 'Mar', 4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'Aug', 9: 'Sep',
      10: 'Oct', 11: 'Nov', 12: 'Dec' }

uber['MONTH'] = uber['MONTH*'].map(dic)
```

```python
uber.head()
```

```python
uber["MONTH"].value_counts().plot(kind = "bar", color = "red", figsize=(12,5))
mp.title("Distribution of trips according to Month")
mp.xlabel("MONTH")
mp.xticks(fontsize = 12)
mp.ylabel("NUMBER OF TRIPS")
mp.show()
```

```python
uber['MONTH'].value_counts().plot(kind='pie',autopct='%.2f')
mp.title("")
mp.show()
```

```python
uber['CATEGORY'].value_counts().plot(kind='pie',autopct='%.2f')
mp.grid()
mp.show()
```

```python
uber[["CATEGORY","MILES"]].groupby("CATEGORY").sum().plot(kind = "bar", color = "green", figsize=(6,3))
mp.title("Miles Wise Category")
mp.xlabel("CATEGORY")
mp.ylabel("NUMBER OF MILES TRAVEL")
mp.show()
```

```python
sns.catplot(x = "MONTH*", kind = "box", data = uber)
```

```python
pd.crosstab(uber.MONTH,uber.PURPOSE).plot(kind = "area", stacked=False, figsize=(16,12))
mp.title("Month wise Purpose")
mp.xlabel("MONTH")
mp.ylabel("Number of Trips")
mp.show()
```

```python
sns.heatmap(uber.corr(),annot=True)
mp.show()
```
